##**Desafio 6 - Rendimento Ações**##

Você está trabalhando num banco e preciso
informar ao seu cliente em que ano foi a maior rentabilidade de dividendo
pago por ação da empresa Johnson&Jonson:

1.Agregar os rendimentos de um mesmo ano

2.Calcular a valorização do rendimento de um ano para outro

3.Informar o ano de maior rentabilidade

Código do dataset: JohnsonJohnson

Dicas:

• Criar uma nova representação dos dados, onde a informação é anual

• Criar uma nova coluna informando a valorização anual

• numpy.trunc()

• método .shift()

• método .idxmax()

```python
jj = data('JohnsonJohnson')
jj
```

```python
# Questão 1:

jj.rename(columns={'time':'year', 'JohnsonJohnson': 'rendimento'}, inplace=True)
jj['year'] = np.trunc(jj['year']).astype(int)
jj_anual = jj.groupby('year').sum()
jj_anual
```

```python
# Questão 2:

jj_anual['Valorização'] = (jj_anual['rendimento']/jj_anual['rendimento'].shift(1) *100 - 100).round(2)
jj_anual
```

```python
# Questão 3:

ano_max = jj_anual['rendimento'].idxmax()
ano_max
```

**id**: Subject identifier variable.

**y**: Housing status response, coded as (1) for street living, (2) for community living and (3) for independent housing.

*No DF, temos 0 (situação de rua), 1 (vivendo em comunidade) e 2(moradia independente)*

**sec**: Section 8 rent certificate indicator.

```python
# Questão 1:

cm_rua = cm.loc[cm['moradia'] == 0.0]
cm_rua_med = (len(cm_rua)/len(cm) *100)
cm_rua_med

print(f'Em média, {cm_rua_med :.1f}% das pessoas estavam em situação de rua durante as medições.')

```
