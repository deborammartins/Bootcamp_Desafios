## Importações


```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

## Dados

```python
df = pd.read_csv('monitoramento_tempo.csv')
```

```python
df.info()
```

```python
df['data'] = pd.to_datetime(df['data'])
```

```python
df
```

## Gráficos

```python
plt.figure(figsize=(15,8))
plt.plot(df['data'], df['temperatura'])
plt.title('Temperatura no Tempo')
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'green')
eixo.set_title('Temperatura no momento', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=16)
eixo.set_xlabel('Data', fontsize=16)
eixo.legend(['temperatura'], loc = 'lower right', fontsize=15)
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'green', ls='dotted') #linha pontilhada
eixo.set_title('Temperatura no momento', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=16)
eixo.set_xlabel('Data', fontsize=16)
eixo.legend(['temperatura'], loc = 'lower right', fontsize=15)
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'green', lw=1) #Espessura da linha
eixo.set_title('Temperatura no momento', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=16)
eixo.set_xlabel('Data', fontsize=16)
eixo.legend(['temperatura'], loc = 'lower right', fontsize=1)
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'green')

eixo.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1)) #criando um intervalo de datas
eixo.set_title('Temperatura no momento', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=16)
eixo.set_xlabel('Data', fontsize=16)
eixo.legend(['temperatura'], loc = 'lower right', fontsize=1)
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'green', marker='o') #criando um marcador bolinha

eixo.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1)) #criando um intervalo de datas
eixo.set_title('Temperatura no momento', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=16)
eixo.set_xlabel('Data', fontsize=16)
eixo.legend(['temperatura'], loc = 'lower right', fontsize=1) #adicionando legenda
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'green')

eixo.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1)) #criando um intervalo de datas
eixo.set_title('Temperatura no momento', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=16)
eixo.set_xlabel('Data', fontsize=16)
eixo.legend(['temperatura'], loc = 'lower right', fontsize=1)
eixo.grid(True) #adicionando linhas de grade
```

```python
fig = plt.figure(figsize=(15,8))

eixo = fig.add_axes([0, 0, 1, 1])
eixo2 = fig.add_axes([0.7, 0.65, 0.3, 0.3]) #criando um gráfico dentro de outro

eixo.plot(df['data'], df['temperatura'], color = 'green')
eixo.grid(True)
eixo.set_xlim(datetime.date(2014,5,1), datetime.date(2014,6,1)) #focando no mes de Maio
eixo.set_ylim(270,320) #criando intervalo de temperatura
eixo.set_title('Temperatura em Maio/2014', fontsize=20, pad=20)
eixo.legend(['temperatura'], loc = 'lower right', fontsize=12)
eixo.set_ylabel('Temperatura', fontsize=16)
eixo.set_xlabel('Data', fontsize=16)

azul_esquerda = df['data'] < datetime.datetime(2014,5,1) #criando um intervalo de data antes de maio
azul_direita = df['data'] > datetime.datetime(2014,6,1) #criando um intervalo de data depois de maio

eixo2.plot(df['data'], df['temperatura'], color = 'green')
eixo2.plot(df[azul_esquerda]['data'], df[azul_esquerda]['temperatura'], color = 'blue') #destacando em azul as datas antes de maio
eixo2.plot(df[azul_direita]['data'], df[azul_direita]['temperatura'], color = 'blue')  #destacando em azul as datas depois de maio
eixo2.set_xlim(datetime.date(2014,1,1), datetime.date(2015,1,1)) #focando no ano todo
eixo2.set_title('Temperatura em 2014', fontsize=10)
eixo2.legend(['temperatura'], loc = 'lower right', fontsize=8)
eixo2.set_ylabel('Temperatura', fontsize=10)
eixo2.set_xlabel('Data', fontsize=10)
eixo2.grid(True)
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'green')

eixo.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1)) #criando um intervalo de datas
eixo.set_title('Temperatura no momento', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=16)
eixo.set_xlabel('Data', fontsize=16)
eixo.legend(['temperatura'], loc = 'lower right', fontsize=1)
eixo.grid(True) #adicionando linhas de grade

x1 = df['data'][df['temperatura'].idxmax()] #local de origem da seta
y1 = max(df['temperatura'])

x2 = df['data'][df['temperatura'].idxmax() - 7000]
y2 = max(df['temperatura']) - 5

eixo.annotate("Máximo", xy=(x1,y1), fontsize=14, xytext=(x2,y2), arrowprops=dict(arrowstyle="->", facecolor='black'))

eixo.axhline(max(df['temperatura']), color = 'k', linestyle='--')
eixo.axhline(min(df['temperatura']), color = 'k', linestyle='--')
```
